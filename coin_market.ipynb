```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://coinmarketcap.com/ico-calendar/ended/'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
# Initialize lists to store data
project = []
ico_price = []
stage = []
start_date = []
end_date = []
goal = []
```

```python
# Loop through multiple pages (adjust the range as needed)
for page_number in range(1, 14):  # Example: Loop through pages 1 to 13
    url = f'https://coinmarketcap.com/ico-calendar/ended/?page={page_number}'
    page = requests.get(url).text
    soup = BeautifulSoup(page, 'html.parser')
    table = soup.find('table', {'class': 'cmc-table'})

    for row in table.find_all('tr')[1:]:
        columns = row.find_all('td')

        if len(columns) >= 6:
            project.append(columns[0].text.strip())
            ico_price.append(columns[1].text.strip())
            stage.append(columns[2].text.strip())
            start_date.append(columns[3].text.strip())
            end_date.append(columns[4].text.strip())
            goal.append(columns[5].text.strip())
```

```python
# Create a DataFrame from the collected data
data = {
    "Project": project,
    "Ico Price": ico_price,
    "Stage": stage,
    "Start date": start_date,
    "End date": end_date,
    "Goal": goal
}

```

```python
df = pd.DataFrame(data)
df
```

```python
# Export the data to a CSV file
df.to_csv('coins_market.csv', index=False)
```
